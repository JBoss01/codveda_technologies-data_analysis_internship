```python
# ============================
# Level 1 - Task 1: Data Cleaning and Preprocessing
# House Prediction Dataset - Codveda Internship
# ============================

# 1. Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

print("Libraries imported successfully!")
```

```python
# ============================
# 2. Load House Prediction Dataset
# ============================

# Correct filename from your folder
df = pd.read_csv('../../data/raw/4) house Prediction Data Set.csv', 
                 sep='\s+', 
                 header=None, 
                 names=['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 
                        'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV'])

print("House Prediction Dataset Loaded Successfully!")
print(f"Shape: {df.shape}")
df.head()
```

```python
# ============================
# 3. Basic Exploration & Insights
# ============================

print(f"Dataset Shape: {df.shape} (Rows, Columns)")
print("\n=== Column Information ===")
print(df.info())

print("\n=== Statistical Summary ===")
display(df.describe().round(2))

print("\n=== Missing Values ===")
missing = df.isnull().sum()
print(missing[missing > 0])

print("\n=== Duplicate Rows ===")
print(f"Number of duplicate rows: {df.duplicated().sum()}")
```

```python
# ============================
# 4. Data Cleaning
# ============================

df_clean = df.copy()

# 4.1 Handle Missing Values (using median)
df_clean = df_clean.fillna(df_clean.median())

# 4.2 Remove Duplicate Rows
initial_rows = len(df_clean)
df_clean = df_clean.drop_duplicates()
print(f"Removed {initial_rows - len(df_clean)} duplicate rows")

# 4.3 Standardize Data Types
df_clean['CHAS'] = df_clean['CHAS'].astype(int)   # Charles River dummy variable

print("Data Cleaning Completed!")
print(f"Original Shape: {df.shape}")
print(f"Cleaned Shape : {df_clean.shape}")
```

```python
# ============================
# 5. Save Cleaned Dataset
# ============================

import os
os.makedirs('../../data/processed', exist_ok=True)

df_clean.to_csv('../../data/processed/house_cleaned.csv', index=False)

print("Cleaned dataset saved to: data/processed/house_cleaned.csv")
```

```python
# ============================
# Final Cleaning & Save (Fixed)
# ============================

# Use the current cleaned dataframe
if 'df_clean' not in locals():
    df_clean = df.copy()

df_final = df_clean.copy()

# Ensure column names are lowercase
df_final.columns = df_final.columns.str.lower()

# Key columns for outlier detection
key_columns = ['crim', 'rm', 'age', 'dis', 'tax', 'lstat', 'medv']

print("Columns available:", df_final.columns.tolist())

# Outlier Capping
for col in key_columns:
    if col in df_final.columns:
        Q1 = df_final[col].quantile(0.25)
        Q3 = df_final[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        df_final[col] = df_final[col].clip(lower=lower, upper=upper)
        print(f"Outliers capped in {col}")
    else:
        print(f"⚠️ Column {col} not found")

# Save the final version
import os
os.makedirs('../../data/processed', exist_ok=True)

df_final.to_csv('../../data/processed/house_cleaned_final.csv', index=False)

print("\nFinal cleaned dataset saved successfully!")
print(f"Final Shape: {df_final.shape}")
```

```python
# ============================
# 6. Final Insights & Summary
# ============================

print("="*60)
print("HOUSE PRICE PREDICTION - DATA CLEANING SUMMARY")
print("="*60)
print(f"Original Dataset Shape     : {df.shape}")
print(f"Cleaned Dataset Shape      : {df_clean.shape}")
print(f"Total Rows Removed         : {df.shape[0] - df_clean.shape[0]}")
print(f"Missing Values Handled     : Yes (Median Imputation)")
print(f"Duplicates Removed         : Yes")
print(f"Target Variable (MEDV)     : House Price in $1000s")
print("="*60)

# Quick Insight
print(f"\nAverage House Price: ${df_clean['MEDV'].mean():.2f} (in $1000s)")
print(f"Cheapest House Price : ${df_clean['MEDV'].min():.2f} (in $1000s)")
print(f"Most Expensive House : ${df_clean['MEDV'].max():.2f} (in $1000s)")
```
